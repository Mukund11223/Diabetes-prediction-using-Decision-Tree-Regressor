```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
```

```python
print(diabetes['DESCR'])
```

```python
diabetes['target']
```

```python
#independent features 
X=pd.DataFrame(diabetes['data'],columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
X
```

```python
y=diabetes['target']
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#correlation
X_train.corr()
```

```python
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
treeR=DecisionTreeRegressor()
```

```python
treeR.fit(X_train,y_train)
```

```python
y_pred=treeR.predict(X_test)
```

HYPERPARAMETER TUNING 

```python
params={
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(treeR,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
select_model=DecisionTreeRegressor(criterion='squared_error',max_depth=2,max_features='log2',splitter='best')
select_model.fit(X_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(select_model,filled=True)
```
